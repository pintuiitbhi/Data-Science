# <a id ="4">4.  Feature Engineering </a>

Now let's create some **features**.

A *ProtectedCustomer* is one who has both Online backups and security, similarly I create other features

```python
df["ProtectedCustomer"] = ["Yes" if df.OnlineBackup[i]=="Yes" and df.OnlineSecurity[i]=="Yes" else "No" for i in range(len(df))]
```

A *StreamerCustomer* is one who has both StreamingMovies and StreamingTV, similarly I create other features

```python
df["StreamerCustomer"] = ["Yes" if df.StreamingMovies[i]=="Yes" and df.StreamingTV[i]=="Yes" else "No" for i in range(len(df))]
```

A *FamilyCustomer* is one who has partners and dependents.

```python
df["FamilyCustomer"] = ["Yes" if df.Partner[i]=="Yes" or df.Dependents[i]=="Yes" else "No" for i in range(len(df))]
```

A *PowerUser* is one who is protected customer and streamer customer.

```python
df["PowerUser"] = ["Yes" if df.ProtectedCustomer[i]=="Yes" and df.StreamerCustomer[i]=="Yes" \
                   and df.DeviceProtection[i]=="Yes" and df.TechSupport[i]=="Yes" else "No" for i in range(len(df))]
```

*FamilyMultiple* is column which contains "yes" if it is family customer and multipleLines.

```python
df["FamilyMultiple"] = ["Yes" if df.FamilyCustomer[i]=="Yes" and df.MultipleLines[i]=="Yes" else "No" for i in range(len(df))]
```

I created another column - *FullCharges* which is the *product of tenure and monthly charges* and it closely resembles TotalCharges .

"TotalCharges" is "object" type. So convert "TotalCharges" to float type, which will help in calculation. 

```python
df["TotalCharges"].dtypes
```

```python
#convert to float type from object type
df['TotalCharges'] = df['TotalCharges'].replace(r'\s+', np.nan, regex=True)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df["FullCharges"] = df.tenure * df.MonthlyCharges
```

```python
df["Discount"] = df.FullCharges - df.TotalCharges
```

## <a id="5-2"> 5.2 Remove Outliers </a>

```python
cols=[ 'Partner', 'Dependents', 'tenure', 
       'PaperlessBilling', 'MonthlyCharges', 'TotalCharges',
       'FullCharges', 'Discount' ]
```

```python
#Determine outliers in dataset

for i in cols:
    quartile_1,quartile_3 = np.percentile(df[i],[25,75])
    quartile_f,quartile_l = np.percentile(df[i],[1,99])
    IQR = quartile_3-quartile_1
    lower_bound = quartile_1 - (1.5*IQR)
    upper_bound = quartile_3 + (1.5*IQR)
    print(i,lower_bound,upper_bound,quartile_f,quartile_l)

    df[i].loc[df[i] < lower_bound] = quartile_f
    df[i].loc[df[i] > upper_bound] = quartile_l

```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

df1=remove_outlier(df, 'TotalCharges')
df1=remove_outlier(df1, 'MonthlyCharges')
df1=remove_outlier(df1, 'FullCharges')
df1=remove_outlier(df1, 'Discount')
```

## <a id="5-6">5.6 Feature Selection </a>
Recursive Feature Elimination (RFE) is based on the idea to repeatedly construct a model and choose either the best or worst performing feature, setting the feature aside and then repeating the process with the rest of the features. This process is applied until all features in the dataset are exhausted. The goal of RFE is to select features by recursively considering smaller and smaller sets of features.

```python
df.columns.values
```

```python
df_final_vars=df.columns.values.tolist()
y=['Churn']
X=[i for i in df_final_vars if i not in y]
```

```python
logreg = LogisticRegression()
rfe = RFE(logreg)
rfe = rfe.fit(df[X], df[y] )
print(rfe.support_)
print(rfe.ranking_)
```

The feature ranking, such that ranking_[i] corresponds to the ranking position of the i-th feature. Selected (i.e., estimated best) features are assigned rank 1.

```python
#Selected Features
cols= [ 'Partner',      
       'InternetService_Fiber optic', 
       'OnlineSecurity_Yes',
        'OnlineBackup_Yes',
        'DeviceProtection_Yes',
       'MonthlyCharges',
       'StreamingMovies_No internet service', 
       'Contract_One year', 'Contract_Two year',
       'PaymentMethod_Credit card (automatic)',
      'PaymentMethod_Electronic check']
        
X=df[cols]
y=df['Churn']
X.shape
```

 ### <a id="6-1A">A.  [Holdout Method] Logistic regression model </a>

```python
logreg = LogisticRegression(C=0.5) #creating the model
logreg.fit(X_train, Y_train) # training the model.
```

```python
#calculating the training error and its accuracy
y_pred_tra = logreg.predict(X_train)
print('Accuracy of logistic regression classifier on training set: {:.2f}'.format(logreg.score(X_train, Y_train)))
print('Missclassification rate for training set: {:.2f}'.format(np.mean(y_pred_tra != Y_train)))
```

```python
#calculating the validation error and its accuracy
y_pred_val = logreg.predict(X_validation)
print('Accuracy of logistic regression classifier on validation set: {:.2f}'.format(logreg.score(X_validation, Y_validation)))
print('Missclassification rate for validation set: {:.2f}'.format(np.mean(y_pred_val != Y_validation)))
```

**Interpretation of result :** 
As we can see above that training missclassification error (0.21 ) is approximately equal to validation error (0.22) , so the model is not overfitting. 
Hence it generalizes the model and it will work better on unseen data i.e test data. 

Now let's see how model works on test set data.

```python
#calculating test error and model accuracy on test set
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, Y_test)))
print('Missclassification rate for test set: {:.2f}'.format(np.mean(y_pred != Y_test)))
```

```python
from sklearn.metrics import confusion_matrix
# Confusion matrix
from sklearn.metrics import confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
 
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(Y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
class_names = ['Not churned','churned']

plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```

The result is telling us that* we have (808 +187) correct predictions and (160+117) incorrect predictions.*

The precision is the ratio tp / (tp + fp) where tp is the number of true positives and fp the number of false positives. The precision is intuitively the ability of the classifier to not label a sample as positive if it is negative.

The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. The recall is intuitively the ability of the classifier to find all the positive samples.

The F-beta score can be interpreted as a weighted harmonic mean of the precision and recall, where an F-beta score reaches its best value at 1 and worst score at 0.

The F-beta score weights the recall more than the precision by a factor of beta. beta = 1.0 means recall and precision are equally important.

### <a id="6-1C"> C.  [Holdout method] KNN </a>

Created a function for KNN:                                                                                                                                                  

```python
def knnClassifier(X_train,Y_train,num_neighbours_range,X_validation,Y_validation):   
    #Setup arrays to store training and test accuracies
    neighbors = np.arange(1,num_neighbours_range)
    train_accuracy =np.empty(len(neighbors))
    test_accuracy = np.empty(len(neighbors))

    for i,k in enumerate(neighbors):
        #Setup a knn classifier with k neighbors
        knn = KNeighborsClassifier(n_neighbors=k)

        #Fit the model
        knn.fit(X_train, Y_train)

        #Compute accuracy on the training set
        train_accuracy[i] = knn.score(X_train, Y_train)

        #Compute accuracy on the test set
        test_accuracy[i] = knn.score(X_validation, Y_validation) 

    #Generate plot
    plt.figure()
    plt.figure(figsize=(8,8))
    plt.title('k-NN Varying number of neighbors')
    plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
    plt.plot(neighbors, train_accuracy, label='validation accuracy')
    plt.legend()
    plt.xlabel('Number of neighbors')
    plt.ylabel('Accuracy')
    plt.show()
```

 ## <a id="6-3">6.3 Stratified K fold cross-validation </a>

Stratified K-Folds cross-validator Provides train/test indices to split data in train/test sets. This cross-validation object is a variation of KFold that returns stratified folds. The folds are made by preserving the percentage of samples for each class.

The data set is split using Stratified Kfold. In each split model is created and predicted using that model. The final predicted value is average of all model.

### <a id="6-3A"> A. [Stratified K fold cross-validation] Logistic Regression model</a>

```python
stratifiedkfold=[]
def stratifiedKFold (model): 
    kf = model_selection.StratifiedKFold(n_splits=5,shuffle=True)
    pred_test_full =0
    cv_score =[]
    i=1
    for train_index,test_index in kf.split(X_intermediate,Y_intermediate):
        print('{} of StratifiedKFold {}'.format(i,kf.n_splits))
        xtr,xvl = X_intermediate.iloc[train_index],X_intermediate.iloc[test_index]
        ytr,yvl = Y_intermediate.iloc[train_index],Y_intermediate.iloc[test_index]

        #model
       # logreg = LogisticRegression(C=0.3)
        model.fit(xtr,ytr)
        score = roc_auc_score(yvl,model.predict(xvl))
        print('ROC AUC score:{:.2f}'.format(score))
        print('_'*50)
        cv_score.append(score)    
        pred_test = model.predict_proba(X_test)[:,1]
        pred_test_full +=pred_test
        i+=1
       
    print('Mean cv Score : {:.2f}'.format(np.mean(cv_score))) 
    return pred_test_full
      
```

## <a id="6-4">6.4 Bootstrapping </a>

The sample will be selected with replacement using the resample() function. Any rows that were not included in the sample are retrieved and used as the test dataset. Next, a decision tree classifier is fit on the sample and evaluated on the test set, a classification score calculated, and added to a list of scores collected across all the bootstraps.

### <a id="6-4A"> A. [Bootstrap Resampling] Logistic Regression Model </a>

```python
from sklearn.utils import resample
from sklearn.metrics import accuracy_score

bootstra=[]
# load dataset
X_intermediate["Churn"]= Y_intermediate
data = X_intermediate
values = data.values

# configure bootstrap
n_iterations = 10 #number of iterations
n_size = int(len(data) * 1) #sample size
# run bootstrap
stats = list()
```

Created a function to do bootstrapping.                                                                                                                             (click "code" to see the code)

```python
def bootstrap(model_name):
    for i in range(n_iterations):
        # prepare train and test sets
        train = resample(values, n_samples=n_size)
        test = np.array([x for x in values if x.tolist() not in train.tolist()])
        # fit model
        model = model_name
        model.fit(train[:,:-1], train[:,-1])
        # evaluate model
        train_pred = model.predict(train[:,:-1])
        score_train = roc_auc_score(train[:,-1], train_pred)
        test_pred = model.predict(test[:,:-1])
        score_test = roc_auc_score(test[:,-1], test_pred)
        print("{} of iteration {}".format(i,n_iterations-1))
        print("ROC AUC Score for Train : {:.2f}".format(score_train))
        print("ROC AUC Score for Validation : {:.2f}".format(score_test))
        print("-"*50)
        stats.append(score_test)
    print("Average ROC AUC Score for Validation : {:.2f}".format(np.mean(stats))) 
    res= (np.mean(stats))
    # plot scores
    plt.hist(stats)
    plt.show()
    # confidence intervals
    alpha = 0.95
    p = ((1.0-alpha)/2.0) * 100
    lower = max(0.0, np.percentile(stats, p))
    p = (alpha+((1.0-alpha)/2.0)) * 100
    upper = min(1.0, np.percentile(stats, p))
    print('%.1f confidence interval %.1f%% and %.1f%%' % (alpha*100, lower*100, upper*100))
    return res
```

```python
print("[Logistic Regression] The ROC AUC score for each bootstrap iteration: ")
bootstra.append(bootstrap(LogisticRegression()))
```

There is a 95% likelihood that the confidence interval 64.4% and 73.0% covers the true skill of the model.

A histogram of the 1,000 accuracy scores is created showing a Gaussian-like distribution.
